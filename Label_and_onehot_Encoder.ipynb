```python
#importing the libraries
import pandas as pd
import numpy as np

#reading the dataset
df=pd.read_csv("salary.csv")
df
```

```python
df.info()
```

```python
#LabelEncoder
#converting Unique categorical variables into Unique numerical values
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()
df['Country']=label_encoder.fit_transform(df['Country'])
df.head()
```

```python
#OneHotEncoder
# It simply creates additional features based on the number of unique values in the categorical feature
df = pd.get_dummies(df, columns=['Country'], dtype='int')
df
```
